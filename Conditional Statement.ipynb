# Conditional Statement

**What is Conditional Statement?**

Conditional /decision making/ Statements are statements executed based on the condition.

There are four types of conditional statements in python. They are as follows:
- 1. if statement
- 2. if else statement
- 3. Nested if statement
- 4. if elif else statement

    if Statement

- The if statement is used to test a specific condition. If the condition is true, if-block statement will be executed.
- The condition of if statement can be any valid logical expression which can be either evaluated to True or False.

Syntax for if statement:

      if (condition):

            if block statement
    
      next line of code

    if else statement

If the condition result is True, then if block statements will be executed. If the condition result is False, then else block statements will be executed.

```python
a = int(input('enter the age: '))

if a >= 18:
    print('vote')
else:
    print("below 18 can't vote")
```

```python
username = "Giri"

name = input("enter the name: ")

if name == username:
    print('welcome to the facebook')
else:
    print('login credintial is wrong')
```

**What are Logical Operators?**

Logical Operators are operators which do some logical operation on the operands and return True or False.

    Note: In logical operators, False indicates 0(zero) and True indicates non-zero value. Logical operators on boolean types
- and : If both the arguments are True then only the result is True
- or : If at least one argument is True then the result is True
- not : complement of the boolean value

```python
nationality = "INDIAN"
CGPA = 6.31

name = input('Enter Your name: ')
print(name)
Country = input('Enter ur Country: ')
cgpa = float(input('Enter ur CGPA:'))

if Country == nationality and CGPA <= cgpa:
    print('You are Eligible for the Schoolarship')
else:
    print('you are not eligible')
```

    Nested if Statement

- Nested if statement is an if statement inside another if statement. Here, a user can decide among multiple options based on multiple conditions.

```python
#Nested if statement

a = int(input('Enter your age :'))
c = float(input('Enter Your CGPA:'))

if a>17 and a<47:
    if c>9:
        print('accepted')
    else:
        print('Waiting List')
else:
    print('application denined')
```

     if elif else Statement

The conditions are evaluated in the written order. If one condition becomes True then the set of statements associated with that particular condition are executed and the execution comes out without checking other conditions.

```python
marks = int(input('enter ur marks:'))

if marks>95:
    print('S Grade')
elif marks>90:
    print('A Grade')
elif marks>80:
    print('B Garde')
elif marks>70:
    print('C Garde')
elif marks>60:
    print('D Garde')
elif marks>50:
    print('E Garde')
else:
    print('Fail')
    
```

```python
a = int(input("Enter your age\n"))
c = float(input("Enter your GPA\n"))

if a > 17 and a < 45 and c > 9:
    print("Accepted.")
elif a > 17 and a < 45 and c > 7 and c < 9:
    print("Pass for interview.")
else:
    print("Application Denied.")
```

    More Examples on Conditional Statements

```python
a = float(input("Enter 1st No: "))
b = float(input("Enter 2nd No: "))
if a > b:
    print(a,"is the largest No.")
else:
    print(b,"is the largest No.")
```

```python
#if else statement using tertiary operator

a = float(input("Enter 1st No: "))
b = float(input("Enter 2nd No: "))
max = a if a > b else b
print("Largest No.=", max)
```

```python
#if elif else statement

a = float(input("Enter 1st No: "))
b = float(input("Enter 2nd No: "))
if a > b:
    print(a,"is the largest No.")
elif a == b:
    print(a,"and",b,"are equal.")
else:
    print(b,"is the largest No.")
```

```python
#Half Nested if statement 

a = float(input("Enter 1st no: "))
b = float(input("Enter 2nd no: "))
c = float(input("Enter 3rd no: "))
if a > b:
    if a > c:
        print("largest no.= ",a)
    else:
        print("largest no.= ",c)
else:
    print("largest no.= ",b)
```

```python
a = float(input('Enter 1st No: '))
b = float(input('Enter 2nd No: '))
c = float(input('Enter 3rd No: '))
if(a >= b):
    if(a >= c):
        largest = a
    else:
        largest = c
else:
    if(b >= c):
        largest = b
    else:
        largest = c
print("The largest No is",largest)
```

**Variable Swapping**
- Using third variable

```python
# Using third variable

a, b = 10, 20
print("befor swap a:", a,"b: ",b)
temp = a
a = b
b = temp
print("after swap a:",a ,"b: ",b)
```
